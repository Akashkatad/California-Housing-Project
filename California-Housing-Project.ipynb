***Q6.Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.***

```python
# The objective of this question is Create a median calculation function for easy and reusable median value calculations in the dataset.
```

```python
def medain_value(d):
  sorted_data=sorted(d)
  n= len(sorted_data)

  if n % 2 == 0:
    med = (sorted_data[n//2 - 1] + sorted_data[n//2]) // 2

  else:
    med = sorted_data[n//2]
  return  med

s=medain_value(df['total_bedrooms'])
print(s)

# Sort the input data d in ascending order and store it in sorted_data.
# Get the length (number of elements) of the sorted data and store it in n.
# Check if the length n is even.
# If n is even, calculate the median as the average of the two middle values.
# Calculate the median as the middle value of the sorted data.
# Return the calculated median.
# Calculate the median of the "total_bedrooms" column from DataFrame df using the median_value function and store it in variable s.
```

***Q10.Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.***

```python
# The main objective is to classify "total_bedrooms" into size categories (small, medium, large)
# and create a new column, "total_bedroom_size," for better data analysis and insights.
```

```python
def newclom(total_bedroom):
  if total_bedroom <= 10:
    return 'Small'
  elif 10< total_bedroom <1000:
    return 'Medium'
  else:
    return 'large'
df['total_bedroom_size'] = df['total_bedrooms'].apply(newclom)
df.head()

# The code defines a function newclom(total_bedroom) to categorize "total_bedrooms"
# into 'Small', 'Medium', or 'Large'. It creates a new column "total_bedroom_size"
# in DataFrame df using this function. The first 100 rows of the updated DataFrame are displayed for analysis.
```
